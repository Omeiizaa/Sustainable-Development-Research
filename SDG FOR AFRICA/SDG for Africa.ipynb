    FILTERING:

<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    Selecting relevant columns for the analysis since we are focusing on SDG 8 and SDG 9. I then filtered data for 5 African Countries for the purpose of this analysis.

```python
Required_SDGs = ['country', 'year', 'goal_8_score', 'goal_9_score']
df_Required = df[Required_SDGs]
df_Required 
```

```python
selected_countries = ['Nigeria', 'Kenya', 'South Africa', 'Ghana', 'Morocco']
df_selected_africa = df_Required[df_Required['country'].isin(selected_countries)]
```

<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">

```python
descriptive_stats = df_selected_africa.groupby('country')[['goal_8_score', 'goal_9_score']].agg(['mean', 'median', 'std', 'min','max'])
descriptive_stats
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    VISUALIZATION TRENDS:

<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    Visualized trends in SDG 8 and 9 over time for these selected countries

```python
sns.set(style="whitegrid")
plt.figure(figsize=(20, 10))

#For SDG 8
plt.subplot(2,1,1) 
sns.lineplot(data=df_selected_africa, x='year', y='goal_8_score', hue='country', marker='o')
plt.title('Trend of SDG 8 (Economic Growth) over time', fontsize=28, fontweight='bold')
plt.xlabel('Year', fontsize=20, fontweight='bold')
plt.ylabel('SDG 8 score', fontsize=20, fontweight='bold')
plt.legend(title='country')

#For SDG 9
plt.subplot(2,1,2)
sns.lineplot(data=df_selected_africa, x='year', y='goal_9_score', hue='country', marker='o')
plt.title('Trend of SDG 9(Infrastructure development) over time', fontsize=28, fontweight='bold')
plt.xlabel('Year', fontsize=20, fontweight='bold')
plt.ylabel('SDG 9 score', fontsize=20, fontweight='bold')
plt.legend(title='country')
plt.tight_layout()

plt.show()
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    CORRELATION:
</div>
<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
Finding the correlation coefficient of SDG 8 and 9 using Pearson.
</div>

```python
correlation_results = {}

for country, data in df_selected_africa.groupby('country'):
    correlation = data['goal_8_score'].corr(data['goal_9_score'], method='pearson')

    correlation_results[country] = correlation
    print(f"{country} - Correlation between SDG 8 and SDG 9: {correlation:.2f}")
```

```python
# I made the correlation data into a dataframe for easier visualiztion.
correlation_df = pd.DataFrame(list(correlation_results.items()), columns=['country', 'r'])
print("\nCorrelation Coefficients between SDG 8 and SDG 9 scores by Country:")
correlation_df
```

<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
To find any strong correlations
</div>

```python
strong_correlations = correlation_df[correlation_df['r'].abs() >= 0.5]
print("\nCountries with Strong Correlation between SDG 8 and SDG 9 Scores:")
print(strong_correlations)
```

<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
From the above, it can be seen that there is a strong correlation between SDG 8 and 9 both positive and negative as it returned all the coutries. The following heatmap shows the significance level for each country.
</div>

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
    VISUALIZATION: HEATMAP
</div>
<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
Visual of how strong the relationship between SDG 8  and 9 is across the selected countries
</div>

```python
# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_df.set_index('country').T, 
            annot=True, 
            cmap='coolwarm', 
            center=0, 
            cbar=True)
plt.title('Correlation between SDG 8 and SDG 9 Scores by Country', fontsize = 15, fontweight = 'bold')
plt.show()
```
